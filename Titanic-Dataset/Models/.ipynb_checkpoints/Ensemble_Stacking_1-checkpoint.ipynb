# Importing Standard Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing Cleaned data from Cleaner_1

```python
x_train = pd.read_csv('../CleanedData/Cleaner1_X_train', index_col = 0)
```

```python
x_train.head()
```

```python
y_train = pd.read_csv('../CleanedData/Cleaner1_Y_train', index_col = 0)
```

```python
y_train.head()
```

```python
x_test = pd.read_csv('../CleanedData/Cleaner1_X_test', index_col = 0)
```

```python
x_test.head()
```

# Importing all algorithms for stacking

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.ensemble import StackingClassifier
```

```python
estimators = [
    ('lr', LogisticRegression()),
    ('knn', KNeighborsClassifier(n_neighbors = 3)),
    ('mnb', MultinomialNB()),
    ('rfc', RandomForestClassifier(n_estimators = 500)),
    ('svc', SVC(C = 10, gamma = 0.0001, kernel = 'linear'))
]
```

# Fitting

```python
sc = StackingClassifier(estimators = estimators, final_estimator = LogisticRegression())
sc.fit(x_train.drop(['PassengerId'],axis = 1),y_train['Survived'])
```

# Prediction

```python
y_test = sc.predict(x_test.drop(['PassengerId'],axis = 1))
```

# Exporting the predictions

```python
predictions = x_test[['PassengerId']]
```

```python
predictions['Survived'] = y_test
```

```python
predictions.head()
```

```python
predictions.to_csv('../Predictions/Ensemble_Stacking_1', index = False)
```
